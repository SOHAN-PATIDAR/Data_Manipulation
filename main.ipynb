5. write a separate function which take input data as values under the column name "phone"
and processes it to return equivalent ASCII char of the value taken 2 at a time from left to right. Considering the value should be in between 65 to 99. If two-digit number is less than 65 replace it with O (O for Orange)


```python

def to_ascii_helper(phn):
    phn = re.sub('[^0-9]',"",phn)
    #print (phn)
    res = ""
    if len(phn) < 10:  
        for i in range(1,len(phn),2):
            pair = int(phn[i-1]+phn[i])

            if pair <65:
                res+='O'
            else:
                res+=chr(pair)
    else:
        for i in range(1,len(phn)-1,2):
            pair = int(phn[i-1]+phn[i])

            if pair <65:
                res+='O'
            else:
                res+=chr(pair)

    return res 

 
def to_ascii():
    for i in range(len(df)):
        df.loc[i, 'phone'] = to_ascii_helper(df.loc[i,'phone'])
        
    df.rename(columns = {'phone':'coded_phone_number'}, inplace = True)
    # print(df['phone'])

to_ascii()

df

                
```
